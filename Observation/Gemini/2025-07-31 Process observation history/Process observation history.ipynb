```python
import pandas as pd
```

```python
# json files were downloaded from the archive search page at https://archive.gemini.edu/searchform/defaults/

gemini_sequences = pd.concat(
    [
        pd.read_json("Gemini 2024B Observations.json"),
        pd.read_json("Gemini 2025A Observations.json"),
    ],
    ignore_index=True,
)
columns_to_keep = [
    "telescope",
    "instrument",
    "ut_datetime",
    "object",
    "ra",
    "dec",
    "airmass",
    "exposure_time",
]
gemini_sequences = gemini_sequences[columns_to_keep].sort_values(["instrument", "ut_datetime"])
gemini_sequences["ut_date"] = [date[:11] for date in gemini_sequences["ut_datetime"]]
gemini_observations = gemini_sequences[["telescope", "instrument", "object", "ut_date"]].drop_duplicates()
gemini_observations = gemini_observations[gemini_observations["object"].str.startswith("TIC")]
# sorted(gemini_observations["ut_date"].unique())
print(len(gemini_observations))
gemini_observations.columns = ["Telescope", "Instrument", "Target Name", "Date UTC"]
gemini_observations.to_csv("Gemini Observations.csv")
gemini_observations
```
