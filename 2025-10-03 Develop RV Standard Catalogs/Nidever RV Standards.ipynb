```python
from functools import partial
from astroquery.simbad import Simbad
import pandasas pd
```

```python
nidever_table = pd.read_csv("../../Files/Spectroscopic Standards/Nidever et al 2002/Table 1.csv")

nidever_table = nidever_table[nidever_table["Days"] > 100]
nidever_table["ID"] = nidever_table["ID"].str.replace(r"\s+", " ", regex=True)

simbad = Simbad()
simbad.TIMEOUT = 120
simbad.add_votable_fields("ids")
simbad_result = simbad.query_objects(nidever_table["ID"]).to_pandas()

nidever_table
```

```python
def extract_id(id_list, id_name):
    for id in id_list.split("|"):
        if id_name in id:
            return id.replace(id_name, "").strip()
    return ""


simbad_result["HD"] = simbad_result["IDS"].apply(partial(extract_id, id_name="HD"))
simbad_result["TIC"] = simbad_result["IDS"].apply(partial(extract_id, id_name="TIC"))
simbad_result = simbad_result[
    (simbad_result["TIC"] > "") & (simbad_result["HD"] > "") & (simbad_result["RA"] > "") & (simbad_result["DEC"] > "")
].reset_index()

simbad_result.iloc[:15]
```

```python
targets_file = pd.DataFrame()
targets_file["Target Name"] = [f"HD {id}" for id in simbad_result["HD"]]
targets_file["RA"] = simbad_result["RA"].str.replace(" ", ":")
targets_file["DEC"] = simbad_result["DEC"].str.replace(" ", ":")
targets_file["Target Type"] = "RV Standard"
targets_file["Target Source"] = "Nidever 2002 https://arxiv.org/abs/astro-ph/0112477v2"
targets_file.to_csv("../../Data/Targets/Nidever et al 2002.csv", index=False)

targets_file
```

```python
catalog_members = pd.DataFrame()
catalog_members["Target Name"] = [f"HD {id}" for id in simbad_result["HD"]]
catalog_members["Catalog Name"] = "TESS TICv8"
catalog_members["Catalog ID"] = simbad_result["TIC"]
catalog_members.to_csv("../../Data/Catalog Members/Nidever et al 2002.csv", index=False)

catalog_members
```
